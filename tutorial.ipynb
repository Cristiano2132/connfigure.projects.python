# Criando um projeto em Python

1. Crie repositório no GitHub
(Pressupomos aqui que você tenha instalado o git e configurado suas credenciais de acesso na máquina que está trabalhando)



Leia atentamente a página que será exibida após a criação do repositório no GitHub, todas as informações e dicas que você precisa estão lá. Poderemos trazer o repositório para nossa máquina de 3 formas:

2. Repositório local
   
   2.1. Opção 1: Clonar o repositório. Para isso observe na parte superir da imagem ao lado de SSH temos o seguinte endereço `git@github.com:Cristiano2132/tutorial.projects.python.git`. 
      
      * Copie este endereço

      * Abra o terminal do ceu computador
      
      * Navegue até o diretório onde deseja fazer download do seu repositório. Por padrão o terminal abrirá na pasta raiz do computador, você poderá se guiar até o local desejado utilizando o comando `ls`, este comando exibe o nome de arquivos e pastas do diretório atual. Para navegar para uma pasta listada basta utilizar o comando `cd nome_da_pasta`. 
      
      * Clone a o seu repositório. Para isso execite o comando `git clone git@github.com:Cristiano2132/tutorial.projects.python.git`

   2.2. Opção 2: Criar um diretório em sua máquina e então inicializar o git: Para isto basta seguir as instruções que aparecem na imagem em `…or create a new repository on the command line`.
      
      * Abra o prompt e navegue até o diretório
      
      * `git init`

      * `git add README.md`

      * `git commit -m "first commit"`

      * `git branch -M main`

      * `git remote add origin git@github.com:Cristiano2132/tutorial.projects.python.git`

      * `git push -u origin main`

1. Crie o arquivo `.gitignore` e `README.md`
   
   3.1.  `git touch .gitignore`
   
   3.2. `git touch README.md`

2. Abra o VSCode `code .`

3. Edite o arquivo `.gitignore`.

   5.1. abra em seu navegador o site `https://www.toptal.com/developers/gitignore`

   5.2. use as keywords `Python` and `VisualStudioCode` e clique em create. Copie o conteúdo que irá ser exibido a seguir.

   5.3. cole as instruções em seu `.gitignore`

Caso você precise trabalhar com arquivos grandes, csv por exemplo, e deseja mantê-los apenas em seu diretório local, 
você pode fazer isso adicionando-os ao `.gitignore` ou adicionando a 
intrução para ignorar todos os arquivos da respectiva extenção `*.csv`


6. Faça o commit das alterações.
   
   6.1. `git status`. Observe que os arquivos que foram modificados e que não estão no `.gitignore` serão listados
 em `Untracked files`. Além disso o git é seu amigo e já está te dando uma dica do que fazer `nothing added to commit but untracked files present (use "git add" to track)`
 ![](figures/git_status.png)

   6.2. Adicionar os arquivos modificados.
      * `git add .` para adicionar todos de uma vez
      * Adiciionar um arquivo por vez. `git add file_name.lol` para adiciona o arquivo `file_name.lol` por exemplo.

   6.3. Commite as alterações. Após o passo anterior ao executar `git status` novamente observe que os arquivos foram adicionados.
   para commitar as alteraçõs, execute o comando `git commit -m "mesage"`. Observe que podemos adicionar uma mensagem. Existem alguns 
   padrões de mensagem que poderão facilitar a identificação do que está sendo adicionado e comunicação entre o time.
   https://gist.github.com/qoomon/5dfcdf8eec66a051ecd85625518cfd13
      Types.
      
   * API relevant changes

     * `feat` Commits, that adds a new feature

     * `fix` Commits, that fixes a bug

   * `refactor` Commits, that rewrite/restructure your code, however does not change any behaviour

     * `perf` Commits are special `refactor` commits, that improve performance

   * `style` Commits, that do not affect the meaning (white-space, formatting, missing semi-colons, etc)

   * `test` Commits, that add missing tests or correcting existing tests

   * `docs` Commits, that affect documentation only

   * `build` Commits, that affect build components like build tool, ci pipeline, dependencies, project version, ...

   * `ops` Commits, that affect operational components like infrastructure, deployment, backup, recovery, ...

   * `chore` Miscellaneous commits e.g. modifying `.gitignore`
